```python
import numpy as np
import string
```

```python
class Mot():
    

    def _init_(self):
        pass
    
    def concat(self,a,b):
        return np.concatenate((a,b), axis=None)
    
    def accord_fem(self,m,e):
        return self.concat(m,e)
    
    def accord_pluriel(self,m,s):
        return self.concat(m,s)
    
    def rep_vect(self,mot):
        alphabet = list(string.ascii_lowercase)
        vect_mot = np.empty(5,int)
        for i in range(len(mot)):
            for j in range(len(alphabet)):
                if mot[i] == alphabet[j]:
                    vect_mot[i]=j 
        
        return vect_mot
    
    def create_dict(self):
        alphabet = string.ascii_lowercase
        dictionary = dict()
        for k,c in enumerate(alphabet):
            dictionary[c]=k
    
    def rep_vect(self,mot):
        mot = mot.lower()
        n = len(mot)
        vect_mot = np.empty(n,int)
        for i in range(len(mot)):
            vect_mot[i]= dictionary[mot[i]] 
        
        return vect_mot
    
    def rep_text(self,vec):
        n = len(vec)
        mot = ''
        for i in range(n):
            for key, value in dictionary.items():
                if value == vec[i]:
                    #print(key, value)
                    mot= mot+ key 
        
        return mot
    
```
