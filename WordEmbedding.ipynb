```python
import numpy as np
import keras
from keras.layers import Input, LSTM, Embedding, Dense, Activation
from keras.models import Model
from loadEquations import load_equations
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
(inputs,targets) = load_equations("./equations.txt")
#print(len(inputs))
tokenizer = Tokenizer(num_words=200)
tokenizer.fit_on_texts(inputs)
#inputs = tokenizer.texts_to_sequences(inputs)
```

```python
#inputs = pad_sequences(inputs,value=0)
num_tokens = (list(tokenizer.word_index.values())[-1])
targets = np.array(targets)
```

```python
inputs_str = []
for i in range(len(inputs)):
    inputs_str.append(inputs[i].split(' '))
```

```python
sentence_vec = []
for sentence in inputs:
    sentence_vec.append(keras.preprocessing.text.one_hot(text=sentence, n=num_tokens, split=' '))
```

```python
sentence_vec
```

```python
sentence_vec = pad_sequences(sentence_vec,value=0)
sentence_vec = np.array(sentence_vec)
```

```python
sentence_vec
```

```python
print(targets.shape)
print((sentence_vec.shape))
```

```python
input_layer = Input(shape=(sentence_vec.shape[1],))
embed_layer = Embedding(num_tokens + 1,100,input_length=sentence_vec.shape[1])(input_layer)
lstm_layer = LSTM(100, activation='tanh', recurrent_activation='hard_sigmoid', return_sequences=False, go_backwards=True)(embed_layer)
predictions = Dense(1,activation='linear')(lstm_layer)
```

```python
model = Model(inputs=input_layer, outputs=predictions)
model.compile(optimizer='adam',
              loss='MSE',
              metrics=['accuracy'])
print(model.summary())
model.fit(sentence_vec,targets,epochs=5)
```
